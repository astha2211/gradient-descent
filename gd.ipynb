```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
def upload_file():
    from google.colab import files
    uploaded = files.upload()
    df = pd.read_csv(list(uploaded.keys())[0])
    print("Uploaded dataset columns:", df.columns)
    return df
```

```python
def preprocess_data(df):
    x = df[['age', 'affordibility']]
    y = df['bought_insurance']
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
    x_train_scaled = x_train.copy()
    x_test_scaled = x_test.copy()
    x_train_scaled['age'] = x_train_scaled['age'] / 100
    x_test_scaled['age'] = x_test_scaled['age'] / 100
    return x_train_scaled, x_test_scaled, y_train, y_test
```

```python
def create_model():
    model = keras.Sequential([
        keras.layers.Dense(1, input_shape=(2,), activation='sigmoid',
                           kernel_initializer='ones', bias_initializer='zeros')
    ])
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
def train_model(model, x_train, y_train):
    model.fit(x_train, y_train, epochs=500)
    return model
```

```python
def evaluate_model(model, x_test, y_test):
    loss, accuracy = model.evaluate(x_test, y_test)
    return loss, accuracy
```

```python
def get_model_weights(model):
    coef, intercepts = model.get_weights()
    return coef, intercepts
```

```python
def main():
    df = upload_file()
    print(df.head())

    x_train_scaled, x_test_scaled, y_train, y_test = preprocess_data(df)

    model = create_model()
    model = train_model(model, x_train_scaled, y_train)

    loss, accuracy = evaluate_model(model, x_test_scaled, y_test)  # Fixed typo here
    print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")

    prediction = model.predict(x_train_scaled)
    coef, intercepts = get_model_weights(model)
    print(f"Model Coefficients: {coef}, Intercepts: {intercepts}")

    return model, prediction
```

```python
if __name__ == "__main__":
    main()
```
